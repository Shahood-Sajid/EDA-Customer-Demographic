```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from datetime import date
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('D:\\forage\\module2\\CustomerDemographicCleaned.xlsx')
```

```python
df
```

```python
df.info()
```

```python
df.drop('Unnamed: 0',inplace = True,axis = 1)
```

```python
df
```

```python
timefmt = "%Y-%m-%d %H:%M:%S" 
df['DOB'] = pd.to_datetime(df['DOB'], format = timefmt, errors='coerce')
df
```

```python
def from_dob_to_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df['Age'] = df['DOB'].apply(lambda x: from_dob_to_age(x))
```

```python
df.info()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace = True)
```

```python
df = df[df.gender != "Undefined"]
```

```python
df['gender'].value_counts()
```

```python
df['Age']=df['Age'].astype('int64')
```

```python
bins   = [18, 40, 65, 90]
labels = ['Adults (18-40)','Mid Age (40-60)','Elderly (65-90)']
df['age_group'] = pd.cut(df['Age'], bins, labels=labels)
```

```python
df
```

```python
df.info()
```

```python
df['age_group'].value_counts()
```

```python
sns.relplot(x = 'Age', y = 'past_3_years_bike_related_purchases',data = df,kind = 'line', color = 'pink', hue = 'gender', col = 'gender',ci = None )
plt.show()
```

```python
sns.relplot(x = 'Age', y = 'past_3_years_bike_related_purchases',data = df,kind = 'line', color = 'pink', hue = 'gender', col = 'age_group',ci = None)
plt.show()
```

```python
sns.relplot(x = 'age_group', y = 'past_3_years_bike_related_purchases',data = df,kind = 'line', color = 'pink', hue = 'gender', col = 'gender', ci = None)
plt.show()
```

```python
sns.relplot(x='age_group', y='past_3_years_bike_related_purchases',kind='line',data=df, hue='job_industry_category',col='job_industry_category',col_wrap=3,ci=None)
```
